```python
import json
import os
from natsort import natsorted # pip install natsort

# babsNum=[71,72,73,74,75,76]
babsNum=[4]

for i in babsNum:
    filelocation=f"/Users/msh/Developer/Flutter/Mafatih/python/Babs/{i}"
    listFiles = natsorted(os.listdir(filelocation))  # dir is your directory path
    for onefile in listFiles:
        fileFinal={}
        dirr= f"{filelocation}/{onefile}"
        f = open(dirr, "r", encoding='utf8')
        fileRead = json.load(f)
        # try:
        fileData=fileRead[onefile]
        fileData["number"]=int(onefile)
        fileData["tozih"]={2:""}
        fileData["farsi"]={2:""}
        fileData["delay"]=1

        Arabic=""

        for i, data in enumerate(fileData["arabic"].values()):
            Arabic=Arabic+data+f"({i+1})"+" "


        fileData["arabic"]={2:Arabic}
        fileFinal[onefile]=fileData

        fileDes=f"/Users/msh/Developer/Flutter/Mafatih/python/Babs/5/{onefile}"


        with open(fileDes, "w", encoding='utf8') as write_file:
            json.dump(fileFinal, write_file, ensure_ascii=False, indent=4)
        print(onefile)

        # except:
        #     print(f"didnt do {onefile}")
        #     pass
```
